## 1: Importando as bibliotecas

```python
# Importando as bibliotecas necessárias
import pandas as pd
```

## 2: Avaliando os dataframes

```python
# Importando os arquivos CSV para DataFrames do Pandas
df_customers = pd.read_csv("Dataset_Original\\olist_customers_dataset.csv")
df_geolocation = pd.read_csv("Dataset_Original\\olist_geolocation_dataset.csv")
df_items = pd.read_csv("Dataset_Original\\olist_order_items_dataset.csv")
df_payments = pd.read_csv("Dataset_Original\\olist_order_payments_dataset.csv")
df_reviews = pd.read_csv("Dataset_Original\\olist_order_reviews_dataset.csv")
df_orders = pd.read_csv("Dataset_Original\\olist_orders_dataset.csv")
df_products = pd.read_csv("Dataset_Original\\olist_products_dataset.csv")
df_sellers = pd.read_csv("Dataset_Original\\olist_sellers_dataset.csv")
```

```python
# Criando um dicionário para facilitar a análise
dfs = {
    'customers':   df_customers,
    'geolocation': df_geolocation,
    'items':       df_items,
    'payments':    df_payments,
    'reviews':     df_reviews,
    'orders':      df_orders,
    'products':    df_products,
    'sellers':     df_sellers,
}
```

```python
# Criando uma tabela com o resumo de duplicatas e nulos
resumo = pd.DataFrame([
    {
        'dataset':          nome,
        'total_linhas':     len(df),
        'duplicatas':       df.duplicated().sum(),
        'dup_%':            round(df.duplicated().sum() / len(df) * 100, 2),
        'colunas_c_nulo':   (df.isnull().sum() > 0).sum(),
        'total_nulos':      df.isnull().sum().sum(),
        'nulos_%':          round(df.isnull().sum().sum() / df.size * 100, 2)
    }
    for nome, df in dfs.items()
])

resumo
```

## 3: Analisando e limpando o dataframe de geolocalização

```python
# Visualizando as primeiras linhas do dataframe de geolocalização
dfs['geolocation'].head()
```

```python
# Removendo as colunas de latitude e longitude
dfs['geolocation'].drop(columns=['geolocation_lat', 'geolocation_lng'], inplace=True)
```

```python
# Retirando as linhas duplicadas do dataframe de geolocalização
dfs['geolocation'].drop_duplicates(inplace=True)

#Confirmando que não existem mais linhas duplicadas no dataframe de geolocalização.
dfs['geolocation'].duplicated().sum() 
```

## 4: Analisando o dataframe de reviews

```python
# Visualizando as primeiras linhas do dataframe de reviews
dfs['reviews'].head() # Existem informações nulas, mas o que é importante é que a nota da avaliação.
```

```python
# Verificando a quantidade de nulos na coluna de nota da avaliação
dfs['reviews']['review_score'].isnull().sum() # Não existem nulos na coluna de nota da avaliação.
```

## 5: Analisando o dataframe de ordens

```python
# Visualizando as primeiras linhas do dataframe de ordens
dfs['orders'].head()
```

```python
# Verificando colunas com nulos no dataframe de ordens
dfs['orders'].isnull().sum() 
```

```python
# Verificando as linhas que possuem nulos na coluna de data de entrega
dfs['orders'][dfs['orders']['order_delivered_customer_date'].isnull()]
```

```python
# Verificando as linhas que possuem nulos na coluna de data de entrega e que possuem o status de "delivered" (entregue)
dfs['orders'][dfs['orders']['order_delivered_customer_date'].isnull()][dfs['orders']['order_status'].str.contains('delivered')] # Existem 8 ordens que possuem o status de "entregue", mas não possuem a data de entrega.
```

## 6: Analisando o dataframe de produtos

```python
# Visualizando as primeiras linhas do dataframe de produtos
dfs['products'].head(10)
```

```python
# Verificando colunas com nulos no dataframe de produtos
dfs['products'].isnull().sum() 
```

```python
# Verificando as linhas que possuem nulos na coluna de categoria
dfs['products'][dfs['products']['product_category_name'].isnull()] 
```

```python
#Retirando as linhas com nulos na coluna de categoria do dataframe de produtos
dfs['products'].dropna(subset=['product_category_name'], inplace=True)
```

```python
# Confirmando que não existem mais nulos na coluna de categoria
dfs['products'].isnull().sum() 
```

## 7: Exportando os dataframes corrigidos

```python
# Exportando os DataFrames limpos para arquivos Excel
for nome, df in dfs.items():
    df.to_excel(f'Dataset_corrigido/{nome}.xlsx', sheet_name=nome, index=False)
    print(f'✅ {nome}.xlsx exportado!')

print('✅✅ Todos os arquivos exportados!✅✅')
```
